## Irrelevant notebook for sanity check (Validation)

Kept to show how to verify correctness of data generated

```python
import pickle as pk
```

```python
def check_pickle(data):
    # print(type(data['results'][0]['id']))
    urls = set()
    for i in range(len(data['results'])):
        url = data['results'][i]['id'][0]  # Ensure 'id' is indexable
        if url == '':
            print(f"Warning: empty id in data['results'][{i}]['id']")
        urls.add(url)
    if len(urls) != 226:
        print(f"Warning: {len(urls)} unique ids in data['results'], expected 226")

    # print(type(data['results'][0]['language_model_proba_df']))
    # data['results'][0]['language_model_proba_df'].head()

    for i in range(len(data['results'])):
        # Check for no NaN values
        if (data['results'][0]['language_model_proba_df'].isnull().values.any()):
            print(f"Warning: missing values in ['results'][{i}]['language_model_proba_df']")

    # print(type(data['results'][0]['gold']))
    # print(data['results'][0]['gold'])

    for i in range(len(data['results'])):
        if data['results'][i]['gold'] == '':
            print(f"Warning: empty gold at data['results'][{i}]")

    # print(type(data['results'][0]['rationality']))
    # print(data['results'][0]['rationality'])

    for i in range(len(data['results'])):
        text_candidates = data['results'][i]['text_candidates']
        if (text_candidates.isnull().values.any()):
            print(f"Warning: missing values in text_candidates, check data['results'][{i}]")
        # print(f"Length of text_candidates[{i}] dataframe: {len(text_candidates)}")

        if not (text_candidates['id'].nunique() == 1):
            print(f"Warning: 'id' is not unique in text_candidates[{i}].")
        if not (text_candidates['gold'].nunique() == 1):
            print(f"Warning: 'gold' is not unique in text_candidates[{i}].")
        if (text_candidates['text'].nunique() <= 1 or text_candidates['text'].nunique() > 4):
            print(f"Warning: there are {text_candidates['text'].nunique()} different texts (reviews) in text_candidates[{i}]")

        # Check that length of text for each value in 'summary' is more than 5
        summary_lengths = text_candidates['summary'].apply(len)
        if not (summary_lengths > 5).all():
            print(f"Warning: some summaries have 5 or fewer characters in text_candidates[{i}].")
            print(text_candidates['summary'][summary_lengths <= 5])

        # Check that all values in 'summary' are of type string
        summary_types = text_candidates['summary'].apply(type)
        if not (summary_types == str).all():
            print(f"Warning: Some summaries are not of type string in text_candidates[{i}].")
            print(text_candidates['summary'][summary_types != str])
```
