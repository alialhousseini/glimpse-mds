### RSA SUMMARY COMPUTATION

```python
import os
import subprocess
import sys
```

```python
candidates = "D:\\Universita\\Progetto NLP\\model_evaluation\\NLP-Project\\finalCandidates" #in this folder keep the csv with the candidates summaries


try:
    rsa_command = [
        "python",
        "glimpse/src/compute_rsa.py",
        "--summaries_folder",
        candidates,
        "--model_name",
        ""
    ]
    rsa_scores = subprocess.check_output(rsa_command).decode("utf-8").strip().split("\n")[-1]
    print(f"Computed RSA SUMMARIES")
except subprocess.CalledProcessError:
    print("Failed to compute RSA summaries.")
    sys.exit(1)
```
